# Problem Statement
*The goal is to develop a machine learning model that can accurately predict whether a person is likely to have diabetes based on diagnostic measurements such as pregnancies, glucose level, diastolic blood pressure, triceps skinfold thickness, insulin level, BMI (Body Mass Index), Diabetes Pedigree Function (DPF), and age. This predictive model will help healthcare professionals identify at-risk individuals early and provide timely interventions to improve patient outcomes.*

# *Attribute Description*

| Attribute       | Meaning                                                                                                                  |
| --------------- | ------------------------------------------------------------------------------------------------------------------------ |
| **pregnancies** | Number of times the patient has been pregnant. Higher counts may be linked to diabetes risk for women.                   |
| **glucose**     | Plasma glucose concentration (mg/dL). Higher levels indicate higher blood sugar, a key indicator for diabetes.           |
| **diastolic**   | Diastolic blood pressure (mm Hg). Elevated blood pressure can be related to diabetes and other health risks.             |
| **triceps**     | Triceps skin fold thickness (mm). Measures body fat; higher fat can increase diabetes risk.                              |
| **insulin**     | 2-hour serum insulin (mu U/ml). Shows how much insulin is in the blood. Abnormal levels can indicate diabetes.           |
| **bmi**         | Body Mass Index = weight (kg) / height (m). Obesity (high BMI) is strongly linked to diabetes.                         |
| **dpf**         | Diabetes Pedigree Function. Represents genetic likelihood of diabetes based on family history. Higher means higher risk. |
| **age**         | Patients age in years. Older age groups tend to have a higher risk of developing diabetes.                              |
| **diabetes**    | Target variable. 1 means the patient is diabetic, 0 means non-diabetic. (Not used in clustering directly).               |

# *Handling Outliers*

```python
columns = ['pregnancies', 'glucose', 'diastolic', 'triceps', 'insulin', 'bmi', 'dpf',
           'age', 'diabetes']
fig,ax = plt.subplots(nrows=5,ncols=2,figsize=(12,18))

for idx in range(len(columns)):
    axis = ax.flatten()[idx]
    sns.boxplot(data=df, x=columns[idx], color='#66b3ff', ax=axis)
    axis.set_title(f'{columns[idx]}', fontweight='bold')

plt.tight_layout()
plt.show()
```

```python
def iqr_capping(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    df[column] = np.where(df[column] < lower_bound, lower_bound,
                 np.where(df[column] > upper_bound, upper_bound, df[column]))

    return df
```

```python
columns = ['pregnancies', 'glucose', 'diastolic', 'triceps', 'insulin', 'bmi', 'dpf',
           'age']
for col in columns:
  df = iqr_capping(df,col)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(x,y, test_size = 0.2, stratify=y, random_state=2)
```

# *Training the Model with SVM*

```python
from sklearn import svm
from sklearn.metrics import accuracy_score
```

```python
classifier = svm.SVC(kernel='linear')
```

```python
classifier.fit(X_train, Y_train)
```

```python
X_train_prediction = classifier.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
X_test_prediction = classifier.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy score of the test data : ', test_data_accuracy)
```
